```python
# 1분퀴즈 4
# 입력값 a 가 50보다 크다면
# a를 3으로 나누어 0이면 "A" 출력
# a를 3으로 나누어 1이면 "B" 출력
# a를 3으로 나누어 2이면 "C" 출력
# 입력값 a 가 50이하이면
# a를 2로 나누어 나머지가 0 이면 "D" 출력
# a를 2로 나누어 나머지가 1 이면 "E" 출력
a = int(input('숫자 입력 : '))
if a > 50 :
    if a % 3 == 0 :
        print('A')
    elif a % 3 == 1 :
        print('B')
    else :
        print('C')
else :
    if a % 2 == 0 :
        print('D')
    else :
        print('E')
```

```python
# 1번 - 1
# 기침여부와 체온을 입력받아 기침을 하고 체온이 38.5도 이상이면 "독감일 수 있습니다." 출력
#  (and , or) 사용
caugh = input('기침합니까? ')
fever = float(input('체온은 몇 도입니까? '))

if caugh == 'Y' and fever >= 38.5 :
    print('독감일 수 있습니다.')
```

```python
# 3번 - 1
# 점수를 입력받아서 0점 아래나 100점 초과 입력시 "잘못된 숫자를 입력했습니다." 출력
# 80 ~ 100  사이 입력 -> "A등급" 출력
# 40 ~ 79 사이 입력 -> "B등급" 출력
# 0 ~ 39 사이 입력 -> "C등급" 출력
score = int(input('점수: '))

if score < 0 or score > 100 :
    print('잘못된 숫자를 입력했습니다.')
elif score >= 80 :
    print('A등급')
elif score >= 40 :
    print('B등급')
else :
    print('C등급')
```

```python
#입력받은 나이에 따라 놀이동산 입장료를 출력한다
# 입장료 정가 : 2만원,  1세~6세:무료, 6세~60세:정가, 60세이상 : 정가의 50%
# 이외의 나이를 입력할 경우 오류 메세지 출력
age = int(input('나이 입력 : '))
price = 20000

if 1 <= age < 6 :
    print('입장료 :  무료')
elif 6 <= age < 60 :
    print(f'입장료 : {price}원')
elif age >= 60 :
    print(f'입장료 : {price*.5:.0f}원')
else :
    print('1살 이상의 나이를 입력해 주세요')
```
