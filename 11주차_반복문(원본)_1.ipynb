#### 5.1.3  반복문으로 간단한 프로그램 만들기

```python
# 1부터 50까지 숫자의 합을 구하는 프로그램
result = 0

for i in range(1, 51) :
    result += i

print(result)
```

```python
# result 변수를 반복문 내부에 만들면
for i in range(1, 51) :
    result = 0
    result += i
print(result)
```

```python
# print()를 반복문 내부에 작성하면
result = 0

for i in range(1, 51) :
    result += i
    print(result)
```

```python
result = 0
for i in range(101, 151) :
    result += i
print(result)
```

```python
result = 1
for i in range(1, 51) :
    result *= i
print(result)
```

```python
import random

print('구구단을 외자~ 구구단을 외자~')

for i in range(10) :
    a, b = random.randint(2,9), random.randint(1,9)
    result = a * b
    question = str(a)+'x'+str(b)+'? '
    user = int(input(question))
    if result == user :
        print("정답")
    else :
        print("오답")
```
