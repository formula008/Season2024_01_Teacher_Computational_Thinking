#### 3.2.4  변수를 사용해 간단한 프로그램 만들기 2

image.png

```python
# 1번. 마트에서 과자, 라면, 햄 구매하기
# 과자 - 1200원, 라면 - 800원, 햄 - 2400원
s_cnt = int(input('과자 몇 개를 구매하시겠습니까? '))
r_cnt = int(input('라면 몇 개를 구매하시겠습니까? '))
h_cnt = int(input('햄 몇 개를 구매하시겠습니까? '))

price = 1200*s_cnt + 800*r_cnt + 2400*h_cnt
print('총 금액 :', price)
```

```python
# 2번. 마트에서 25% 할인 행사
s_cnt = int(input('과자 몇 개를 구매하시겠습니까? '))
r_cnt = int(input('라면 몇 개를 구매하시겠습니까? '))
h_cnt = int(input('햄 몇 개를 구매하시겠습니까? '))

price = 1200*s_cnt + 800*r_cnt + 2400*h_cnt
print('할인 전 금액 :', price, ', 할인 후 금액', price * 0.75)
```

```python
# 1번 - 국어,영어,수학 점수를 입력하여 총점과 평균을 구하고 평균은 소숫점 둘째 짜리 까지 구하여라
kor = int(input('국어점수는? '))
mat = int(input('수학점수는? '))
eng = int(input('영어점수는? '))
total = kor + mat + eng
avg = total/3

print(f'총점 : {total}, 평균 : {avg:.2f}')
```
