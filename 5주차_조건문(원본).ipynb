## 4.3  조건이 여러 개일 때: elif
#### 4.3.1  조건이 3개일 때

image.png

```python
# 무인 계산기-2
# 1단계 - 나이 입력받아 입장료 결정하기
age = int(input('나이가 어떻게 되십니까? '))

if age >= 20 :
    print('성인은 15000원입니다.')
    price = 15000
else :
    print('미성년자는 6000원입니다.')
    price = 6000

# 2단계 - 투입금액을 입력받아 입장료와 비교하기
price_input = int(input('금액을 투입해주세요. --> '))


if price_input < price :
    print(price-price_input,'원 더 내셔야 합니다.')
elif price_input == price :
    print('감사합니다. 즐거운 시간 되세요!')
else :
    print(price_input-price,'원 거슬러 드리겠습니다.')
```

```python
# 1분퀴즈 4
# 입력값 a 가 50보다 크다면
# a를 3으로 나누어 0이면 "A" 출력
# a를 3으로 나누어 1이면 "B" 출력
# a를 3으로 나누어 2이면 "C" 출력
# 입력값 a 가 50이하이면
# a를 2로 나누어 나머지가 0 이면 "D" 출력
# a를 2로 나누어 나머지가 1 이면 "E" 출력
a = int(input('숫자 입력 : '))
if a > 50 :
    if a % 3 == 0 :
        print('A')
    elif a % 3 == 1 :
        print('B')
    else :
        print('C')
else :
    if a % 2 == 0 :
        print('D')
    else :
        print('E')
```

```python
# 1번 - 1
# 기침여부와 체온을 입력받아 기침을 하고 체온이 38.5도 이상이면 "독감일 수 있습니다." 출력
#  (and , or) 사용
caugh = input('기침합니까? ')
fever = float(input('체온은 몇 도입니까? '))

if caugh == 'Y' and fever >= 38.5 :
    print('독감일 수 있습니다.')
```

```python
# 2번
# 메뉴는 "김밥" 과 "라면"이 있고 메뉴를 입력받으면
# 김밥 - 야채김밥 : 2500원, 참치김밥 : 3500원
# 라면 - 기본라면 : 3500원, 떡라면 : 4000원, 만두라면 : 4000원
# 다른메뉴 입력시 - "올바른 메뉴를 입력해 주세요" <- 출력
menu = input('어떤 메뉴가 궁금하신가요? ')

if menu == '김밥' :
    print('야채김밥: 2500원')
    print('참치김밥: 3500원')
elif menu == '라면' :
    print('기본라면: 3500원')
    print('떡라면: 4000원')
    print('만두라면: 4000원')
else :
    print('올바른 메뉴를 입력해 주세요')
```

```python
# 3번 - 1
# 점수를 입력받아서 0점 아래나 100점 초과 입력시 "잘못된 숫자를 입력했습니다." 출력
# 80 ~ 100  사이 입력 -> "A등급" 출력
# 40 ~ 79 사이 입력 -> "B등급" 출력
# 0 ~ 39 사이 입력 -> "C등급" 출력
score = int(input('점수: '))

if score < 0 or score > 100 :
    print('잘못된 숫자를 입력했습니다.')
elif score >= 80 :
    print('A등급')
elif score >= 40 :
    print('B등급')
else :
    print('C등급')
```
