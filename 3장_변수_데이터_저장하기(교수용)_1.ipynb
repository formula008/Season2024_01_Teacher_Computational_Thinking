 ## 3번 자동판매기에 토입한 돈과 물건 값이 아래와 같이 변수에저장되었다고 가정하고 잔돈을 계산하여 동전으로 거슬러주는 프로그램을 작성하여라

```python
money = 1600   #투입한돈
price = 1000   # 물건값

change = money - price
print('거스름돈 : ',change,'원')

print('500원 동전 : ',change//500,'개')
print('100원 동전 : ',change%500,'개')
```
