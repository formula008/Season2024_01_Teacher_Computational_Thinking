## 6.2  제어구조로 리스트 다루기
#### 6.2.1  조건문으로 리스트 다루기

image.png

```python
food =['피자', '치킨', '스테이크', '된장찌개']
search = input('찾는 음식은? ')

if search in food :
    print(food.index(search)+1,'위입니다.')
else :
    print(search,'은/는 순위에 없습니다')
```

#### 6.2.2  중첩 제어구조로 리스트 다루기

image.png

```python
import random

# 초기 상태 표로 보여주기
score = []
for i in range(25) :
    score.append(random.randint(0,100))
    print(score)
```

```python
import random

score = []
for i in range(25) :
    score.append(random.randint(0,100))
print(score)

print('리스트를 사용해 70점 이상인 점수 출력하기')
for s in score :
    if s > 70 :
        print(score.index(s) + 1, '번', s)

print('인덱스를 사용해 70점 이상인 점수 출력하기')
for i in range(25) :
    if score[i] > 70 :
        print(i+1,'번', score[i])
```

```python
#1번 - 주민번호 앞6자리와 뒷자리 1개를 입력받아 생년월일과 성별을 출력하세요
jumin = input('주민번호 앞 6잘와 뒷자리 1개 입력(XXXXXX-X) ->')
year = jumin[:2]
year = '20' + year
month = jumin[2:4]
day = jumin[4:6]

if jumin[-1:] == '3' :
    sex = '남자'
else :
    sex = '여자'

print(f'생년월일 : {year}년 {month}월 {day}일 성별은 {sex}입니다.')
```

```python
# 1번 - 1
#아래 book_name 리스트 변수에 책 이름이 저장되어 있다
# book_price 리스트에 책의 가격을 정하는데 15000원~25000원까지 랜덤으로 가격을 정하시오
import random

book_name=['역사탐험대','파이썬','학습법','영단어','여행에세이','삼국지']
book_price = []

for i in range(len(book_name)) :
    book_price.append(random.randint(15000, 25000))

print(book_name)
print(book_price)
```

```python
# 1번 - 2
# 위 문제에서 가장 비싼 책의 이름은 무엇인지 출력하시오
expensive = max(book_price)
idx = book_price.index(expensive)
print('가장 비싼 책은', book_name[idx])
```
