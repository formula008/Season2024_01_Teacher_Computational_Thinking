## 5.4  열심히 달리는 반복문에 브레이크 밟기: break
#### 5.4.1  반복 멈추기

```python
for i in range(20) :
    print(i)
    break
```

```python
for i in range(20) :
    print(i)
    if i == 2 :
        break
```

```python
answer = int(input('숫자 입력 : '))

for i in range(20) :
    print(i, end=' ')
    if i == answer :
        break
```

```python
answer = int(input('숫자 입력 : '))

for i in range(20) :
    if i == answer :
        break
    print(i, end=' ')
```

#### 5.4.2  break를 사용한 게임 만들기

image.png

```python
# 구구단을 외자 - 오답일때 프로그램종료
import random

print('구구단을 외자~ 구구단을 외자~')

while True :
    dan, num = random.randint(2,9),random.randint(1,9)
    user = int(input(f'{dan} X {num} = '))
    if user == dan*num :
        print('정답')
    else :
        print(f'땡!! 정답은 {dan*num}')
        print('프로그램을 종료합니다.')
        break
```

```python
# 구구단을 외자 - 게임을 계속 할것인지를 물어서 할 경루
import random

print('구구단을 외자~ 구구단을 외자~')

while True :
    dan, num = random.randint(2,9),random.randint(1,9)
    user = int(input(f'{dan} X {num} = '))
    if user == dan*num :
        print('정답')
    else :
        print(f'땡!! 정답은 {dan*num}')

    exit = input('계속 하시겠습니까?(y/n) --> ')

    if exit != 'y' :
        break

print('프로그램을 종료합니다.')
```

```python
#종합실습 - 컴퓨터가 생각한 숫자(1~20까지) 맞추기 게임
import random

cnt = 0
com = random.randint(1,20)
print('<< 컴퓨터가 생각한 1~20 숫자 맞추기>>')

while True :
    player = int(input('숫자입력 : '))
    cnt += 1
    if player == com :
        print(f'정답 : {com} --> {cnt}번 만에 맞추었네요' )
        break ;
    elif player > com :
        print('더 작은숫자 입력!')
    else :
        print('더 큰숫자 입력!')
```
